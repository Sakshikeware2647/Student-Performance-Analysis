# Data Visualization

## Histogram of Final Grades

This graph shows the distribution of students' final grades.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(data['G3'], bins=10)

plt.title('Distribution of Final Grades')

plt.xlabel('Final Grade')

plt.ylabel('Number of Students')

plt.show()
```

### Interpretation

The histogram shows that most students scored between 8 and 14 marks, while only a few students scored very high marks.

## Scatter Plot: Study Time vs Final Grade

This graph shows the relationship between study time and final grades.

```python
plt.figure(figsize=(8,5))

plt.scatter(data['studytime'], data['G3'])

plt.title('Study Time vs Final Grade')

plt.xlabel('Study Time')

plt.ylabel('Final Grade')

plt.show()
```

### Interpretation

The scatter plot indicates that students who study more generally tend to achieve better final grades.

## Bar Chart: Average Final Grade by Gender

This graph compares the average grades of male and female students.

```python
gender_average = data.groupby('sex')['G3'].mean()

plt.figure(figsize=(6,5))

gender_average.plot(kind='bar')

plt.title('Average Final Grade by Gender')

plt.xlabel('Gender')

plt.ylabel('Average Grade')

plt.show()
```

### Interpretation

The bar chart shows the average performance of male and female students and helps identify which gender performed better.
